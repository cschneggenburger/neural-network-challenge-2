## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
X_data = ["Education",
          "Age",
          "DistanceFromHome",
          "JobSatisfaction",
          "OverTime",
          "StockOptionLevel",
          "WorkLifeBalance",
          "YearsAtCompany",
          "YearsSinceLastPromotion",
          "NumCompaniesWorked"]


# Create X_df using your selected columns
X_df = attrition_df[X_data]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)

display(X_train)
display(X_test)
display(y_train)
display(y_test)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
ot_encoder = OneHotEncoder(sparse_output=False)

#Fit the OverTime column on X_train
X_train_encoded = ot_encoder.fit_transform(np.array(X_train[['OverTime']]).reshape(-1,1))
```

```python
#Transform training data
X_train_ohe_df = pd.DataFrame(X_train_encoded, columns=ot_encoder.get_feature_names_out())

#concat with original data
X_train_data = pd.concat([X_train.reset_index(drop=True), X_train_ohe_df.reset_index(drop=True)], axis=1)
X_train_data = X_train_data.drop('OverTime', axis=1)


display(X_train_ohe_df)
display(X_train)
display(X_train_data)
```

```python
X_test_encoded = ot_encoder.transform(np.array(X_test['OverTime']).reshape(-1, 1))
X_test_ohe_df = pd.DataFrame(X_test_encoded, columns=ot_encoder.get_feature_names_out())

# X_test_data = pd.concat([X_test, X_test_ohe_df], axis=1).drop(['OverTime'], axis=1)
X_test_data = pd.concat([X_test.reset_index(drop=True), X_test_ohe_df.reset_index(drop=True)], axis=1)
X_test_data = X_test_data.drop('OverTime', axis=1)

display(X_test)
display(X_test_data)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit(X_train_data[["Education",
          "Age",
          "DistanceFromHome",
          "JobSatisfaction",
          "StockOptionLevel",
          "WorkLifeBalance",
          "YearsAtCompany",
          "YearsSinceLastPromotion",
          "NumCompaniesWorked"]])

#Create a DataFrame of the X_train scaled data
X_train_data[["Education",
          "Age",
          "DistanceFromHome",
          "JobSatisfaction",
          "StockOptionLevel",
          "WorkLifeBalance",
          "YearsAtCompany",
          "YearsSinceLastPromotion",
          "NumCompaniesWorked"]] = scaler.transform(X_train_data[["Education",
                                                      "Age",
                                                      "DistanceFromHome",
                                                      "JobSatisfaction",
                                                      "StockOptionLevel",
                                                      "WorkLifeBalance",
                                                      "YearsAtCompany",
                                                      "YearsSinceLastPromotion",
                                                      "NumCompaniesWorked"]])


#Replace X_test data with scaled data
X_test_data[["Education",
          "Age",
          "DistanceFromHome",
          "JobSatisfaction",
          "StockOptionLevel",
          "WorkLifeBalance",
          "YearsAtCompany",
          "YearsSinceLastPromotion",
          "NumCompaniesWorked"]] = scaler.transform(X_test_data[["Education",
                                                      "Age",
                                                      "DistanceFromHome",
                                                      "JobSatisfaction",
                                                      "StockOptionLevel",
                                                      "WorkLifeBalance",
                                                      "YearsAtCompany",
                                                      "YearsSinceLastPromotion",
                                                      "NumCompaniesWorked"]])

#Check the resulting DF's
display(X_train_data)
display(X_test_data)
```

```python
# Create a OneHotEncoder for the Department and Attrition columns
dep_encoder = OneHotEncoder(sparse_output=False)
att_encoder = OneHotEncoder(sparse_output=False)


# Fit the encoder to the training data
dep_encoder.fit(y_train[['Department']])
att_encoder.fit(y_train[['Attrition']])

# Create four new variables by applying the encoder
# to the training and testing data
y_train_dep = dep_encoder.transform(y_train[['Department']])
y_train_dep_columns = dep_encoder.get_feature_names_out(['Department'])
y_train_dep_df = pd.DataFrame(y_train_dep, columns=y_train_dep_columns)
y_test_dep = dep_encoder.transform(y_test[['Department']])
y_test_dep_columns = dep_encoder.get_feature_names_out(['Department'])
y_test_dep_df = pd.DataFrame(y_test_dep, columns=y_test_dep_columns)

y_train_att = att_encoder.transform(y_train[['Attrition']])
y_train_att_columns = att_encoder.get_feature_names_out(['Attrition'])
y_train_att_df = pd.DataFrame(y_train_att, columns=y_train_att_columns)
y_test_att = att_encoder.transform(y_test[['Attrition']])
y_test_att_columns = att_encoder.get_feature_names_out(['Attrition'])
y_test_att_df = pd.DataFrame(y_test_att, columns=y_test_att_columns)

display(y_train_dep_df)
display(y_train_att_df)

display(y_test_dep_df)
display(y_test_att_df)
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
print(f"{len(X_train_data.columns)}")

# Create the input layer
input_layer = layers.Input(shape=(X_train_data.shape[1],), name="input_features")

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu', name='shared_layer1')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu', name='shared_layer2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dep_hidden = layers.Dense(32, activation='relu', name='dep_hidden')(shared_layer2)

# Create the output layer
dep_output = layers.Dense(3, activation='softmax', name='dep_output')(dep_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
att_hidden = layers.Dense(32, activation='relu', name='att_hidden')(shared_layer2)

# Create the output layer
att_output = layers.Dense(2, activation='sigmoid', name='att_output')(att_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dep_output, att_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'dep_output': 'categorical_crossentropy',
                    'att_output': 'categorical_crossentropy'},
              metrics={'dep_output': 'accuracy',
                       'att_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_data, {'dep_output': y_train_dep_df,
                    'att_output': y_train_att_df},
          epochs=100,
          batch_size=32
          )

```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_data, {'dep_output': y_test_dep_df, 'att_output': y_test_att_df})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Department predictions accuracy: {test_results[3]}")
print(f"Attrition predictions accuracy: {test_results[4]}")
```
